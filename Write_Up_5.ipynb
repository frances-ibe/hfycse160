# 5 Results
Before reading the results it is important to understand the terminology that was used in our analysis. The zhvi feature from the Zillow dataset stands for Zillow Home Value Index, which is the median house value based on house value estimates made using Zillow’s proprietary machine learning software. The price point feature from the Yelp dataset is binned into 1 dollar sign, 2 dollar sign, 3 dollar sign, 4 dollar sign, which going forward will be referred to as price point 1, 2, 3, and 4. The price points are bins for the cost of an average meal per individual. Price point 1 is below 10 USD, price point 2 is 11-31 USD, price point 3 is 31-60 USD, and price point 4 is above 61 USD.

We found that there is a strong positive correlation between the average household income and zhvi, which suggests that zip codes with higher average household income tend to have higher zhvi, which suggests there is collinearity between the two variables. This confirms that our choice of zhvi and average household income as two proxies for socioeconomic class status is legitimate, as they both hold to the same trends.

The moderate correlation between average restaurant price point and mean income also suggests that zip codes with higher mean household income on average have more restaurants at higher price points. This suggests, that as zip code regions have a increase in the mean income, whether that be wealthy outliers generating a positive skew in the distribution of income or an increase in the general middle class population household income, there tend to be a better market or at least a more saturated market for higher price point restaurants.

It is interesting to note that our other proxy for socioeconomic standing of a zip code, zhvi, did not show any significant correlation with average restaurant price point. Thus, the feature space where the strong correlation between zhvi and mean income lies is independent of the feature space where the correlation between mean income and average restaurant price point lies.

## 5.3 Research Question 2
The ability to predict the average restaurant price in a zip code region would indicate that consumers make spending decisions based upon the predictive metrics chosen. Initially we hoped to train a multiple regression model to predict average restaurant price in a zip code region using both income and zillow home value index as predictors. However, examination of the scatter plot and the pearson’s correlation coefficient between income and zillow home value index shows relatively strong collinearity between the two predictors. As collinearity is a recipe for poor regression models, we chose to compare two simple linear regression models, predicting average restaurant price from income and Zillow home value index separately. Scatter plots of the true data with the linear regression line of best fit can be found in figures 4 and 5. Based on the regression lines there are negligible positive relationships between average restaurant prices and zillow home value index and average income, respectively. As indicated by the R-squared metric, both regression models capture less the 30 percent of the variation in the average restaurant price data (R-squared=0.288 for Income vs avg price, R-squared=0.037 for zillow home value index vs avg price). Since the two linear regression models fail to capture significant variation within the average price data, we fail to predict average price directly from from the two predictors. Although this appears to indicate that income and zillow home value index are not indicators of consumer buying power, it only indicates that across zip code averages these metrics do not directly correlate with zip code average restaurant price. This does not eliminate the ability to examine variations with zip codes to attempt to understand how consumers inform purchasing choices.

### 5.4.2 Chi-Square Analysis
Several chi-square tests of independence were used to first filter our data to contain only zip code regions that met our independence assumptions and then to analyze the distribution of stars and restaurants across price point levels for each zip code.

Our statistical analysis and models thus far have relied upon four significant assumptions:

Number of restaurants varies within each price point level across all zip codes (number of restaurants is not independent of price point level for the summary of all zip codes). 
This assumption is important because it allows us to analyze restaurant number and proportions for different price points separately as it suggests that there are, in fact, differences between each price level. This allows us to use price point level as an indicator of customer expense, which we use as an indicator of buying power (a customer being able to spend more at a restaurant would indicate higher buying power).Number of restaurants varies within each price point level within each zip code (number of restaurants is not independent of price point level for each zip code). Similar to assumption (1), this assumption allows us to analyze restaurant numbers and proportions for different price points separately, however rather than looking at summaries across all zip codes, this allows us to analyze each zip code separately. In particular, this allows us to calculate and interpret separate correlations for each price point level for proportion of restaurants and zhvi or income.
Number of stars varies within each price point level across all zip codes (number of stars is not independent of price point level for the summary of all zip codes). This assumption is important in a similar way to assumption (1) though it allows us to instead analyze star number and proportions for different price points separately.
Number of stars varies within each price point level within each zip code (number of stars is not independent of price point level for each zip code). This assumption is similar to assumption (2) though it instead allows us to analyze star number and proportions for different price points and different zip codes separately. In particular, this allows us to calculate and interpret separate correlations for each price point level for average number of stars and zhvi or income.

To validate these four assumptions, four chi-square tests were performed, one for each assumption. For validation of assumptions (1) and (3), small chi-square tests was performed on the summary measures across all zip codes. In order to validate assumptions (2) and (4), chi-square tests were performed for each zip code. The setups and results for each of these four model-validating chi-square tests are summarized below:

(1) Observed: number of restaurants in a price level for the summary of all zip codes. 
Expected: total number of restaurants * 0.25
This choice for observed and expected assumes that the total number of restaurants is equally distributed among each price level. 
Results: P-value = 0.0 < 0.01, meaning that the total number of restaurants in each level is not independent of the price point level.

(2) Observed: number of restaurants in a price level for a single zip code. 
Expected: number of restaurants in a zip code * 0.25
This choice of observed and expected assumes that the number of restaurants in a zip code is equally distributed among price levels. 
Results: All p-values were < 0.01 = alpha, except the p-value for the zip code regions 89138, 89166, and 89179 which had p-values of 0.0633, 0.0226, and 0.0572 respectively. For all 33 other zip codes, the number of restaurants per price point level is independent of the price point level. 

(3) Observed: number of stars in a price level for the summary of all zip codes.
Expected: total number of stars * 0.25
This choice for observed and expected assumes that the total number of stars is equally distributed among each price level.
Results: P-value = 0.0 < 0.01, meaning that the total number of restaurants in each level is not independent of the price point level.

(4) Observed: number of stars in a price level for a single zip code. 
Expected: total number of stars in a zip code * 0.25
This choice for observed and expected assumes that the number of stars in a zip code is equally distributed among each price level. 
Results: All p-values were < 0.01 = alpha, except the p-value for the zip code region 89179. For all 35 other zip code regions, the number of stars per price point level is independent of the price point level.

After chi-square tests were used to validate analysis assumptions, two chi-square tests were performed to analyze the interplay between restaurant and star distribution among price point levels. These tests assume that the distribution of stars across price point levels depends on the distribution of restaurants across price point levels. Having a statistically significant p-value for these tests would indicate that the distribution of stars differs from the distribution of restaurants for each price point level. This would suggest that stars are not equally distributed to each restaurant in each price point level, that is, price point levels differences in amount of stars is not only due to the factor of number of restaurants. This could suggest that customers favor certain price point levels, if, for instance, the restaurants within a certain price point level had a higher proportion of stars than those within a different price point level. To examine the influence of restaurant distribution on star distribution for each price point level, the following chi-square tests were performed:

(1) Observed: number of stars in a price level for the summary of all zip codes.
Expected: total number of restaurants per price level divided by total number of restaurants * total number of stars 
Results: P-value = 2.122x10^-9 < 0.01=alpha, meaning that the distribution of stars differs from the distribution of restaurants across price point levels for the summary of the data.

(2) Star Distribution by Restaurant
Observed: number of stars in a price level for a single zip code. 
Expected: total number of restaurants per price level divided by total number of restaurants * total number of stars in the zip code
Results: All p-values were < 0.01 = alpha, except the p-values for the zip code regions 89138, 89166, 89169, and 89179. Notably, all of these zip-codes, except for 89169 also failed the chi-square test of independence for number of restaurants across price point levels within a zip code. For all 32 other restaurants, the distribution of stars differs from the distribution of restaurants per across price point levels for each zip code.

To analyze differences between zip codes and the general trend for the whole city, two more chi-square tests were performed to analyze whether the proportion of stars and restaurants per price level in each zip code region varied from the total proportion of stars and restaurants for all zip codes. The setup of the tests is described below, and the results for each zip code are summarized in Table 4. 

(1) Restaurant Distribution by Total
Observed: number of restaurants in a price level for a single zip code.
Expected: total number of restaurants per price level divided by total number of restaurants * total number of restaurants in the zip code

This choice for observed and expected assumes that the distribution of restaurants per price level is the same for the summary of all zip codes and all individual zip codes. Any zip codes that have a statistically significant p-value differ significantly from the summary restaurant distribution, suggesting that such zip codes should not be reduced to summary measures but instead the zip codes should be analyzed separately. 

(2) Star Distribution by Total
Observed: number of stars in a price level for a single zip code.
Expected: total number of stars per price level divided by total number of stars * total number of stars in the zip code
This choice for observed and expected assumes that the distribution of stars per price level is the same for the summary of all zip codes and all individual zip codes. Any zip codes that have a statistically significant p-value differ significantly from the summary restaurant distribution, suggesting that such zip codes should not be reduced to summary measures but instead the zip codes should be analyzed separately. 



*Table 3: Results from Star Distribution by Restaurant, Restaurant Distribution by Total, and Star Distribution by Total chi-square tests. Whether or not the p-value computed by the test was less than alpha = 0.01 is displayed for each zip code, as well as the total number of restaurants in each zip code.*

From examination of Table 3, it is worth note that the zip code regions with the smallest number of restaurants were the ones who failed all three of the displayed chi-square tests. Namely, the zip codes which failed all three tests were 89138, 89166, and 89179, which had 7, 7, and 2 restaurants total. The sample size of such zip code regions was quite small, and, as such, perhaps they should be removed from the analysis, as the data presented from these zip code regions is not a representative random sample of the zip code region.

Further examining the results of these three chi-square tests, it is worth note all zip code regions that failed either the Star Distribution by Restaurant or the Star Distribution by Total chi-square tests also failed the Restaurant Distribution by Total chi-square test, the test which the largest number of zip code regions failed. In total, 20 zip code regions did not have a p-value less than 0.01 for the Restaurant Distribution by Total chi-square test. Comparing this to the Star Distribution by Total chi-square test, for which only 5 zip code regions did not have a p-value less than 0.01, one could attribute such variation to the claim that restaurant distribution across price point levels within each zip code is, in general, more similar to the city-wide restaurant distribution than per zip code star distribution is to city-wide star distribution. This would mean that different zip code regions see greater star distribution variation from city-wide star distribution, which suggests that certain zip code regions favor different price point levels than the city-wide trend in price point level favorization. Such a conclusion further validates our analysis of different zip code regions as different areas and gives insight that there are, in fact, significant differences between each zip code.
